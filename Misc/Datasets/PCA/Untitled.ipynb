# PCA(Principal Component Analysis) 

# 1. load Python Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# 2. Create the data

```python
def print_mean_std(s):
    print(s.mean())
    print(s.std())
```

```python
x = np.random.normal(200,10,500)
plt.hist(x)
plt.show()
print_mean_std(x)
```

```python
y = 0.5*x + np.random.normal(10,5,500)
plt.hist(y)
plt.show()
print_mean_std(y)
```

# Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```

# covariance - co-relation

```python
co_mat=np.cov(x,y)
co_mat
```

```python
sns.heatmap(co_mat,annot=True)
```

```python
correlation_matrix = np.corrcoef(x,y)
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True)
```

# PCA

```python
from numpy.linalg import eig
eig_val ,eig_vects = eig(co_mat)
```

```python
eig_val
```

```python
eig_vects
```

```python
v1=eig_vects[0]
v2=eig_vects[1]
```

```python
np.dot(v1,v2)
```

```python
eig_val_per = eig_val /np.sum(eig_val)
eig_val_per
```

# Create new data

```python
z=np.vstack([x,y]).T
z
```

```python
new_data = np.dot(z,eig_vects)
new_data
```

```python
df =pd.DataFrame(new_data,columns=["PC1","PC2"])
df
```

```python
co_mat=np.cov(df["PC1"],df["PC2"])
co_mat
```

```python
plt.scatter(df["PC1"],df["PC2"])
```

```python
plt.scatter(x,y)
plt.scatter(df["PC1"],df["PC2"])
```
